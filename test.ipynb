```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("fifa21 raw data v2.csv")
data.head(20)
data
myData = pd.DataFrame(data)
```

```python
for index,value in myData["Height"].items():
    myData.at[index,"Height"] = myData.at[index,"Height"][:-2]
for index,value in myData["Weight"].items():
    myData.at[index,"Weight"] = myData.at[index,"Weight"][:-2]
print(myData["Height"])
print(myData["Weight"])
```

```python
for index,value in myData["Club"].items():
    myData.at[index,"Club"] = myData.at[index,"Club"][4:]
myData["Club"]
```

```python
print(myData["Joined"])
date = pd.Timestamp.now().date()
date = str(date)
for index,value in myData["Joined"].items():
    if (int(date[:4])- int(myData.at[index,"Joined"][-4:]))>10:
        print("more")
    else:
        print("less")
```

```python
array = myData["Value"].to_numpy()
index = np.where(array == '‚Ç¨0')
for index,value in myData["Value"].items():
    figure = myData.at[index,"Value"][1:-1]
    for number in np.nditer(index):
        if number == index:
            myData.at[index,"Value"]=0
    if "." in figure:
        figure=np.float(figure)*1000
        myData.at[index,"Value"] = figure
    elif figure == '':
        pass
    else:
        figure=np.int(figure)*1000
        myData.at[index,"Value"] = figure
myData.at[289,"Value"]



```
